# Data : 1 'House Rental Data'

### 1. Load rental data with first column as index column

```python
house_rental = pd.read_csv('https://raw.githubusercontent.com/edyoda/data-science-complete-tutorial/master/Data/house_rental_data.csv.txt',index_col='Unnamed: 0')
```

### 2. Find all houses which has more than 2 bedroom and price less than 50k

```python
house_rental[(house_rental.Bedroom > 2) & (house_rental.Price < 50000)].head()
```

### 3. Find all houses which has more than 3 bedroom and price less than 50k. Update their price with 60k

```python
house_rental[(house_rental.Bedroom > 3) & (house_rental.Price < 50000)].head()
```

```python
def f(r):
    if r.Bedroom > 3 and r.Price < 50000:
        return 60000
    else:
        return r.Price
house_rental['New_Price'] = house_rental.apply(f,axis = 1)
```

```python
house_rental[house_rental.Bedroom > 3].head()
```

### 2. Find all the male who is consuming alcohal more than 5 years and has diabetes result positive.

```python
health_care_data.rename(columns={'Duration of alcohol consumption(years)':'alcohol_consumption_duration','Diabetes Result':'diabetes'},inplace=True)
```

```python
health_care_data[(health_care_data.alcohol_consumption_duration > 5) & (health_care_data.diabetes == 'YES')].head()
```

### 3. Find where more people consume alcohol in rural or urban?

```python
health_care_data.rename(columns={'Place(location where the patient lives)':'patient_location'},inplace=True)
```

```python
health_care_data.patient_location.value_counts()
```

More people consume alcohol in rural

### 4. Find average age of male and female who consume alcohol.

```python
health_care_data.Gender.unique()
```

```python
def f(r):
    if r.Gender == 'male':
        return 'Male'
    if r.Gender == 'female' or r.Gender == 'female ':
        return 'Female'
    if r.Gender == 'transgender':
        return 'TransGender'
health_care_data['gender'] = health_care_data.apply(f,axis = 1)
```

```python
health_care_data.gender.value_counts()
```

```python
health_care_data.groupby('gender').Age.mean()
```

### 5. Display distribution for different type of alcohol.

```python
health_care_data.rename(columns={'Type of alcohol consumed':'consumed_alcohol_type'},inplace=True)
```

```python
health_care_data.consumed_alcohol_type.value_counts()
```

```python
def f(r):
    if r.consumed_alcohol_type == ' branded liquor':
        return 'branded liquor'
    else:
        return r.consumed_alcohol_type
    
health_care_data['consumed_alcohol_type'] = health_care_data.apply(f,axis=1)
```

```python
health_care_data.consumed_alcohol_type.value_counts()
```

### 6. Display average age of perople who has diabetes positive.

```python
health_care_data[health_care_data.diabetes == 'YES'].Age.mean()
```

### 7. Find is there any missing value of age in given data.

```python
health_care_data[['Age']].info()
```

There are no missing values of age in given data

### 8. If gender is male and age is less than 35 then change diabetes result to NO .

```python
def diabetes(r):
    if r.gender == 'Male' and r.Age < 35:
        return 'NO'
    else:
        return r.diabetes
    
    
health_care_data['diabetes'] = health_care_data.apply(diabetes,axis = 1)
```

### 9. Find all record where Place is missing.

```python
health_care_data[health_care_data.patient_location.isnull()].head()
```

### 10. Change place to urban if gender is female else rural.

```python
health_care_data[health_care_data.gender == 'Female'].head()
```

```python
def place(r):
    if r.gender == 'Female':
        return 'urban'
    else:
        return 'rural'
health_care_data['patient_location'] = health_care_data.apply(place,axis=1)
```
